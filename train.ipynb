```python
from dataloader import *

from matplotlib import pyplot as plt
import numpy as np
```

```python
transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Resize(size=(224, 224))
            ])
            
dataloader = get_loader(
    root_folder="data/flickr8k/images/",
    annotation_file="data/flickr8k/captions.txt",
    transform=transform
)
```

```python
def show_example(dataloader, num_examples=16):
    i = 0
    img_id = 0
    for idx, (imgs, targets) in enumerate(dataloader):

        plt.imshow(np.transpose(imgs[img_id], (1, 2, 0)))
        t = ''
        vector = []
        for row in range(len(targets)):
            vector += [targets[row, img_id].item()]
            
        for ltr in vector:
            word = dataloader.dataset.vocab.itos[ltr]
            if word == '<PAD>':
                break
            t += word + ' '
        plt.title(t)
        plt.show()
        i += 1
        if i == num_examples:
            break

show_example(dataloader)
```
